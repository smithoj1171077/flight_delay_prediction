```python
import os
BASE_PATH = "../src/data/2023-us-civil-flights-delay-meteo-and-aircraft/versions/9/"
```

### 1. geolocations

```python
GEOLOCATIONS = os.path.join(BASE_PATH,"airports_geolocation.csv")
```

```python
import pandas as pd

geo_df = pd.read_csv(GEOLOCATIONS)

```

```python
len(geo_df)
```

```python
geo_df.head(2)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

lats = geo_df["LATITUDE"]
lons = geo_df["LONGITUDE"]

def plot_USA():
    fig = plt.figure(figsize=(8, 6))
    ax = plt.axes(projection=ccrs.PlateCarree())

    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=":")
    ax.add_feature(cfeature.STATES, linestyle=":")

    ax.set_extent([-125, -66.5, 24, 49], crs=ccrs.PlateCarree())

    ax.scatter(lons, lats, color="red", s=50, transform=ccrs.PlateCarree())

    plt.show()
    
def plot_world():
    fig = plt.figure(figsize=(8, 6))
    ax = plt.axes(projection=ccrs.PlateCarree())

    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=":")
    ax.add_feature(cfeature.STATES, linestyle=":")

    ax.scatter(lons, lats, color="red", s=50, transform=ccrs.PlateCarree())

    plt.show()
```

```python
plot_world()
```

```python
plot_USA()
```
